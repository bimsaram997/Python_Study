```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
X,y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
# Extract the column at index 2
X = X[:, 2]
print(X.shape)

# Reshape to 2D array
X = X.reshape((-1,1))
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
mean_BMI = X_train.mean()
std_BMI = X_train.std()

print("Mean BMI (Training Data):", mean_BMI)
print("Standard Deviation of BMI (Training Data):", std_BMI)
```

```python
y_pred = model.predict(X_test)
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

```python
# Raw BMI value to predict
raw_BMI = 27.5

# Replace these with the actual mean and std of the training BMI data
mean_BMI = -0.00039086669665966856 
std_BMI = 0.047535846816921815  

# Scale the BMI
scaled_BMI = (raw_BMI - mean_BMI) / std_BMI
scaled_BMI = np.array([[scaled_BMI]])  # Reshape to 2D array for model input

# Predict disease progression
predicted_progression = model.predict(scaled_BMI)
print(f"Predicted Disease Progression for BMI {raw_BMI}: {predicted_progression[0]}")
```
